```python
from matplotlib import pyplot
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
pyplot.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
pyplot.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

pyplot.scatter('a', 'b', c='c', s='d', data=data)
pyplot.xlabel('entry a')
pyplot.ylabel('entry b')
pyplot.show()
```

```python
mydict = {
    'x': np.arange(10),
    'y':  np.arange(10),
    'color': [1, 2, 3, 4, 10, 6, 7, 8, 9, 10],
}

pyplot.scatter('x', 'y', c='color', data=mydict)
pyplot.show()
```
